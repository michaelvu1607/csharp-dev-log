FILTERING | FILTERING | FILTERING | FILTERING | FILTERING | FILTERING

```python
df = pd.read_csv("pokemon.csv")
```

```python
tall_poke = df[df["Height"] >= 2]
print(tall_poke)
legendary = df[df["Legendary"] == True]
print(legendary)
```

```python
# OR -> "|"
water_poke = df[(df["Type1"] == "Water") | (df["Type2"] == "Water")]
print(water_poke)
```

```python
# AND -> "&"
```

```python
ff_poke = df[(df["Type1"] == "Fire") & (df["Type2"] == "Flying")]
print(ff_poke)
```

DATA CLEANING | DATA CLEANING | DATA CLEANING | DATA CLEANING | DATA CLEANING | DATA CLEANING

```python
# Drop irrelevant columns
df = pd.read_csv("pokemon.csv")
df = df.drop(columns=["Legendary", "No"])
print(df)
```

```python
# 1. Handle missing data ("dropna" = "drop not available or NaN")
# ex. within Type2, drop any rows missing a value
df = pd.read_csv("pokemon.csv")
df = df.dropna(subset=["Type2"])
print(df.to_string())
```

```python
# 2. Handle missing data ("fillna")
# ex. replace any missing values with "None"
df = pd.read_csv("pokemon.csv")
df = df.fillna({"Type2": "None"})
print(df)
```

```python
# Fix inconsistent values
# ex. in column "Type1", change all instances of "grass" to "GRASS
df = pd.read_csv("pokemon.csv")
df["Type1"] = df["Type1"].replace({"Grass":"GRASS",
                                  "Fire":"FIRE",
                                  "Water":"WATER"})
print(df.to_string())
```

```python
# Standardize text
df = pd.read_csv("pokemon.csv")
df["Name"] = df["Name"].str.upper()
print(df.to_string())
```

```python
# Fix data types
df = pd.read_csv("pokemon.csv")
df["Legendary"] = df["Legendary"].astype(bool)
print(df.to_string())
```
