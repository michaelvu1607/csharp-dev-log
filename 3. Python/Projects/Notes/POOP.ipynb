```python
# under "class Dog:" are the operations(methods) that can be performed by that class
    # methods can take arguments just like functions
class Dog:
    # init is called anytime a new object of that class is created and takes the arguments simultaneously
    def __init__(self, name, age):
        # self.name creates an attribute of the class Dog
            # stores name as attribute of the object permanently
            # can be referenced later on even in methods inside of that class
        self.name = name
        self.age = age
        # demonstrates referencing in methods
    def get_name(self):
        return self.name

    def get_age(self):
        return self.age

    # methods can modify attributes
    def set_age(self, age):
        self.age = age

    def bark(self):
        print("bark")

    def add_one(self, x):
        return x + 1

# d is an object of class Dog
d = Dog("Bob", 43)
d.bark()
print(d.add_one(5))
print(type(d))
print(d.name)

# demonstrates permanence of attributes
d2 = Dog("Bill", 51)
print(d2.name)

# calling referencing in methods
print(d.get_name())
print(d. get_age())
print(d2.get_name())
print(d2. get_age())

# calls a method that modifies age; then has age printed
d.set_age(23)
print(d.get_age())
```

```python
# Multiple Classes
class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade # 0-100
    def get_name(self):
        return self.name
    def get_age(self):
        return self.age
    def get_grade(self):
        return self.grade

class Course:
    def __init__(self, name, max_students):
        self.name = name
        self.max_students = max_students
        self.students = []
    def add_student(self, student):
        if len(self.students) < self.max_students:
            self.students.append(student)
            return True
        else:
            return False
    def get_avg_grade(self):
        value = 0
        for student in self.students:
            value += student.get_grade()
        return value / len(self.students)

s1 = Student("Tim", 19, 95)
s2 = Student("Bill", 19, 75)
s3 = Student("Jill", 19, 65)

course = Course("Science", 2)
course.add_student(s1)
course.add_student(s2)
course.add_student(s3)
print(course.get_avg_grade())
```

```python
# Inheritance
# Pet is a parent class
class Pet:
    def __init__(self, name, age):
        self.name = name
        self.age = str(age)
    def show(self):
        print(f"I am {self.name} and am {self.age} years old")
    def speak(self):
        print("idk what to say")

# Cat and Dog are derived/child classes
class Cat(Pet):
    def __init__(self, name, age, color):
        # super allows child class to call methods from parent class
        super().__init__(name, age)
        self.color = color
    def speak(self):
        print("meow")
    def show(self):
        print(f"I am {self.name} and am {self.age} years old and {self.color}")

class Dog(Pet):
    def speak(self):
        print("bark")

p = Pet("Bob", "7")
p.show()
c = Cat("Bill", 15, "brown")
c.show()
c.speak()
d = Dog("Jill", 21)
d.show()
d.speak()
```
