```python
"""
1.
The Restaurant Order System
Goal: Create two classes where one class handles a collection of objects from the other.

The Dish Class
Attributes: name (string) and price (float).

Method: __init__ to set these two values.

The Order Class
Attributes: table_number (int) and items (a list that starts empty).

Methods:

add_dish(dish_obj): This method should take an instance of the Dish class and append it to the items list.

calculate_total(): This should loop through the items list, add up all the prices, and return the total.

print_receipt(): This should print the table number and the names of all dishes ordered.
"""

class Dish:
    def __init__(self, name, price):
        self.name = name
        self.price = float(price)

class Order:
    def __init__(self, table_number):
        self.dishes = []
        self.table_number = table_number
    def add_dish(self, dish_obj):
        self.dishes.append(dish_obj)
    def calculate_total(self):
        total = 0
        for item in self.dishes:
            total += item.price
        return total
    def print_receipt(self):
        print(self.table_number)
        for item in self.dishes:
            print(f"{item.name} : {item.price}")

dish = Dish("hotdog", 1)
order = Order("1")
order.add_dish(dish)
print(order.print_receipt())
```

```python
"""
2.
The Zoo Management System
Goal: Create a Zoo that can hold different Animal objects and tell you which ones are hungry.

The Animal Class
__init__ attributes: name, species, and is_hungry (a boolean).

The Zoo Class
__init__ attribute: animals (a list that starts empty).

Methods:

add_animal(animal_obj): Adds an Animal instance to the list.

show_hungry_animals():

This should loop through the list.

If an animal's is_hungry attribute is True, print: "The [species] named [name] needs food!"

feed_all():

This should loop through the list and set every animal's is_hungry attribute to False.
"""

class Animal:
    def __init__(self, name, species, is_hungry):
        self.name = name
        self.species = species
        self.is_hungry = bool(is_hungry)

class Zoo:
    def __init__(self):
        self.animals = []
    def add_animal(self, animal_obj):
        self.animals.append(animal_obj)
    def show_hungry_animals(self):
        for animal in self.animals:
            if animal.is_hungry:
                print(f"The {animal.species} named {animal.name} needs food!")
    def feed_all(self):
        for animal in self.animals:
            if animal.is_hungry:
                animal.is_hungry = False

a = Animal("Bill", "cat", True)
zoo = Zoo()
zoo.add_animal(a)
zoo.show_hungry_animals()
```

```python
"""
3.
The Music Playlist Manager
Goal: Create a system where a Playlist can find a specific Song and "play" it.

The Song Class
Attributes: title, artist, and play_count (starts at 0).

Method: play() — This should simply increase play_count by 1.

The Playlist Class
Attributes: name and songs (a list).

Methods:

add_song(song_obj): Adds a song to the list.

find_and_play(song_title):

This should loop through the songs.

If the song's title matches the song_title passed into the method, call that song's .play() method and print "Now playing [title]".

Bonus: If the loop finishes and no song was found, print "Song not found."
"""

class Song:
    def __init__(self, title, artist):
        self.title = title
        self.artist = artist
        self.play_count = 0
    def play(self):
        self.play_count += 1

class Playlist:
    def __init__(self, name):
        self.name = name
        self.songs = []
    def add_song(self, song_obj):
        self.songs.append(song_obj)
    def find_and_play(self, song_title, song_artist):
        for song in self.songs:
            if song.title == song_title:
                song.play()
                print(f'Now playing "{song_title}" by {song_artist}')
            else:
                print("Song not found.")

s1 = Song("Fly Me to the Moon", "Frank Sinatra")
p1 = Playlist("Best Classics")
p1.add_song(s1)
p1.find_and_play("Fly Me to the Moon", "Frank Sinatra")
```

```python
"""
4.
The Inventory Manager
Goal: Create a system where a Warehouse can track Product quantities and "restock" them when they run low.

The Product Class
Attributes: name, price, and stock_quantity.

Method: update_stock(amount) — This should add the amount to the stock_quantity. (Note: amount could be positive for a delivery or negative for a sale).

The Warehouse Class
Attributes: location and inventory (a list).

Methods:

new_product(product_obj): Adds the whole product object to the inventory list.

check_low_stock(threshold):

Loop through the inventory.

If a product's stock_quantity is less than the threshold, print: "Product [name] is low! Only [quantity] left."

restock_product(product_name, amount):

Loop through the inventory to find the product that matches the product_name.

If found, call that product's update_stock(amount) method.
"""

class Product:
    def __init__(self, name, price, stock_quantity):
        self.name = name
        self.price = price
        self.stock_quantity = stock_quantity
    def update_stock(self,amount):
        self.stock_quantity += amount

class Warehouse:
    def __init__(self, location):
        self.location = location
        self.inventory = []
    def new_product(self, product_obj):
        self.inventory.append(product_obj)
    def check_low_stock(self, threshold):
        for item in self.inventory:
            if item.stock_quantity < threshold:
                print(f"Product {item.name} is low! Only {item.stock_quantity} is left. ")
            else:
                print(f"{item.stock_quantity} in stock. ")
    def restock_product(self, product_name, amount):
        for item in self.inventory:
            if product_name == item.name:
                item.update_stock(amount)
                print("Restocking. ")

p1 = Product("rice", 5.00, 10)
w1 = Warehouse("Detroit")
w1.new_product(p1)
w1.check_low_stock(11)
w1.restock_product("rice", 2)
w1.check_low_stock(11)
```
