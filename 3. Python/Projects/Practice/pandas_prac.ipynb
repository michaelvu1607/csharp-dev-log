```python
"""
The Scenario:
You've received a slightly messy version of the Pokémon dataset. You need to clean it up and find some quick statistical insights.
-----
Your Task:
Using the skills from the "Data Cleaning" and "Aggregate Functions" sections of the tutorial, write a script to do the following:
Handle Missing Data: Imagine your DataFrame has a column called Type 2 where some values are missing (NaN). Use fillna() to replace all missing values in that column with the string "None".
Standardize Text: The Name column has names in various cases. Use the .str.lower() method to make every Pokémon name lowercase.
Calculate Statistics: * Find the mean (average) of the Weight column.
Find the max value of the Height column.
Group Analysis: Use groupby() on the Type 1 column and calculate the mean height for each type.
"""

import numpy as np # Only used here to create the 'empty' spots

data = {
    "Name": ["Bulbasaur", "CHARMANDER", "Squirtle", "PIKACHU", "Gengar"],
    "Type 1": ["Grass", "Fire", "Water", "Electric", "Ghost"],
    "Type 2": ["Poison", np.nan, np.nan, np.nan, "Poison"],
    "Height": [0.7, 0.6, 0.5, 0.4, 1.5],
    "Weight": [6.9, 8.5, 9.0, 6.0, 40.5]
}

df = pd.DataFrame(data)
df = df.fillna({"Type 2": "None"})
df_name = df["Name"].str.lower()
print(df["Weight"].mean())
print(df["Height"].max())
df_type1 = df.groupby("Type 1")
print(df_type1["Height"].mean())
```
